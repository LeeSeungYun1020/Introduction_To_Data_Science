##T1. 생일이 같은 사람이 존재할 확률 계산하기
* $N$명의 사람으로 이뤄진 모임이 있다.
* 모임의 사람들의 생일간에는 특별한 연관성이 없다고 가정한다. 
* 즉 어떤 사람의 생일은 1년 366일 (2월 29일도 포함) 중 하루를 모두 같은 확률 (1/366)로 가진다.

1.   $N=10$인 모임에서 생일이 같은 사람이 존재할 확률은 얼마나 되나?
  - 약 11.664541% = 0.11664541  
2.  $N$이 몇 명 이상(Equal or Greater than)이면 생일이 같은 사람이 존재할 확률이 50% 이상이 될까? 
  - 23명  

필요하다면 scipy.special package의 [perm](https://docs.scipy.org/doc/scipy/reference/generated/scipy.special.perm.html#scipy.special.perm) (순열, Permutation), [comb](https://docs.scipy.org/doc/scipy/reference/generated/scipy.special.comb.html#scipy.special.comb) (조합, Combination) 함수를 사용하라.


```python
import numpy as np
import scipy.special as sp
n = 23#@param {type:"integer"}
#@markdown * 1번 문항: n = 10 대입
#@markdown * 2번 문항: n = 22, 23 대입
#@markdown   - 22는 약 47.475%, 23은 약 50.632%
#@markdown   - 23명 이상이어야 확률이 50% 이상
if n >= 367:
  print("100%")
else:
  empty = sp.perm(366, n, exact=False)/ (366**n)
  p = 1 - empty
  print(f"{p * 100: .6f}%")
```

##T2. 주사위 던지기 확률 계산하기
* 주사위를 100번 던져 6이 나오는 횟수와 관련한 확률을 구하려 한다.
1. 6이 나오는 횟수가 20번 이상일 확률은 얼마인가?
  - 21.974984% = 0.21974984  
1. 6이 나오는 횟수가 5번 이하일 확률은 얼마인가?
  - 0.038492% = 0.00038492  

필요하다면 [scipy.stats](https://docs.scipy.org/doc/scipy/reference/stats.html) package의 확률분포 함수를 사용하라.


```python
import scipy.stats as stats
 
# Binomial Random Variable/Distribution
n, p = 100, 1/6
k = 5#@param {type:"integer"}
condition = "\uC774\uD558" #@param ["\uC774\uC0C1", "\uC774\uD558"]
#@markdown * 1번 문항: k = 20, condition = "이상" 선택
#@markdown * 2번 문항: k = 5, condition = "이하" 선택
rv = stats.binom(n,p)
if condition == "이하":
  print(f"P(X<={k})={rv.cdf(k) * 100:.6f}%")
else:
  print(f"P(X>={k})={(1 - rv.cdf(k) + rv.pmf(k)) * 100:.6f}%")
```

##T3. A형 간염 보균자 확률 계산하기
* 부산대학교 병원은 오랜 기간의 임상 조사와 분석을 통해 한국인의 A형 간염 보균과 관련한 특성을 얻었다.
* 결과에 따르면 한국인 10,000명당 평균 16명이 A형 간염균을 가지고 있는 것으로 파악되었다.
* 부산대학교 병원은 부산대학교 정보컴퓨터공학부에 재학 중인 학생 1,000명에 대해 A형 간염 보균 여부를 확인하는 검사를 시행할 계획이다.

1.  검사 결과 1,000명의 학생 중 A형 간염 보균자가 없을 (0명) 확률은 얼마인가?
  - 20.163798% = 0.20163798
  - 포아송 분포를 이용할 경우: 20.189652% = 0.20189652  
1. 1,000명의 학생 중 A형 간염 보균자의 수가 5명 이상일 확률은 얼마인가?
  - 2.357640% = 0.02357640
  - 포아송 분포를 이용할 경우: 2.368228% = 0.02368228

```python
import numpy as np
import scipy.stats as stats
k = 5#@param {type:"integer"}
condition = "\uC774\uC0C1" #@param ["\uC774\uC0C1", "\uC774\uD558"]
#@markdown * 1번 문항: k=0, condition="이하" 선택
#@markdown * 2번 문항: k=5, condition="이상" 선택
n, p = 1000, 16/10000
if condition == "이상":
  ks = np.arange(k, 1001)
else:
  ks = np.arange(0, k + 1)
rv = stats.binom(n,p)
rv2 = stats.poisson(mu=n*p)
print(f"이항. Pr({ks[0]} <= X <= {ks[-1]})={100*sum(rv.pmf(ks)):.6f}%")
print(f"포아송. Pr({ks[0]} <= X <= {ks[-1]})={100*sum(rv2.pmf(ks)):.6f}%")
```

##T4. 콜센터(Call Center) 통화 시간 길이 확률 계산하기
* 어떤 금융 회사 콜 센터(Call Center)는 오랜 기간의 경험과 기록 분석을 통해 고객 서비스를 위한 통화 시간 길이의 특성을 얻었다
* 분석 결과 고객 서비스를 위한 통화 시간은 서로 독립이며 (Independent)
 그 길이는 평균은 4 분(minute)이었다.  
  
1. 어떤 고객과의 통화가 1분 이내에 끝날 확률은 얼마인가?
  - 22.119922% = 0.22119922  
1. 어떤 고객과의 통화가 8분 이상 걸릴 확률은 얼마인가?
  - 13.533528% = 0.13533528  

```python
import scipy.stats as stats
rv = stats.expon(scale=4)
m = 8#@param {type:"number"}
condition = "\uC774\uC0C1" #@param ["\uC774\uC0C1", "\uBBF8\uB9CC"]
#@markdown * 1번 문항: m=1, condition="미만" 선택
#@markdown * 2번 문항: m=8, condition="이상" 선택
 
if condition == "이상":
  print(f"{(1 - rv.cdf(m)) * 100:.6f}%")
else:
  print(f"{rv.cdf(m) * 100:.6f}%")
```

##T5. IQ 확률 계산하기
*  어떤 [조사](https://www.healthline.com/health/average-iq)에 따르면 한국인의 IQ는 평균이 106이라고 한다
* IQ 값의 정규(Normal/Gaussian) 분포를 따르며 표준 편차는 15 정도인 것으로 파악된다.
 
1. 임의로 선정한 어떤 한국인의 IQ 값이 120 이하일 확률은 얼마인가?
  - 82.467606% = 0.82467606  
2. 임의로 선정한 어떤 한국인의 IQ 값이 135 이상일 확률은 얼마인가?
  - 2.659757% = 0.02659757  
3. IQ 값이 상위 1%인 한국인은 IQ가 얼마보다 높아야 할까? (결과가 정수가 아닐 경우 반올림하여 답을 적어라)
  - 140.895218 $\approx$ 141  
 
scipy.stats package는 분포에 대해 pmf/pdf, cdf 외에도 ppf() - percent point function - 라는 함수를 제공한다. ppf()는 cdf의 역함수이다. 예를 들어 CDF(x) =  0.9 (즉 Pr(X <= x) = 90%)인 x 값은 PPF(0.9)를 통해 구할 수 있다.

```python
iq =  135#@param {type:"integer"}
condition = "\uC774\uC0C1" #@param ["\uC774\uC0C1", "\uC774\uD558"]
#@markdown IQ 값 상위
percent =  1#@param {type:"integer"}
#@markdown * 1번 문항: iq=120, condition="이하" 선택
#@markdown * 2번 문항: id=135, condition="이상" 선택
#@markdown * 3번 문항: (IQ 값 상위)percent=1 대입 (1, 2번 문항과 동시 표시)
#@markdown   - 약 140.895로 소수 첫째 자리에서 반올림하면 141
 
import scipy.stats as stats
# standard normal X ~ N(0,1^2)
rv = stats.norm(loc=106, scale=15)
 
if condition == "이하":
  print(f"Pr(X <= {iq}) = {rv.cdf(iq) * 100:.6f}%")
else:
  print(f"Pr(X >= {iq}) = {(1 - rv.cdf(iq-0.00000000000001)) * 100:.6f}%")
print(f"Pr(X <= {rv.ppf(1 - percent/100):.6f}) = {100 - percent}%")
```
